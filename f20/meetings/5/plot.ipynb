```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# to make font bigger in pdf:
# make original font size bigger
# make original plot size smaller
plt.rcParams["font.size"] = 16
```

Advice
* shrink all figs in Python before saving
* make figsize width consistent in all plots (3 in)
* use same original font size

```python
s = pd.Series(np.random.randint(-3,10,100)).cumsum()
ax = s.plot.line(color="red", figsize=(5,3))
ax.set_ylabel("Cases")
ax.set_xlabel("Days")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
plt.tight_layout()
plt.savefig("cases.pdf")
```

```python
s = pd.Series(np.random.randint(-3,10,100)).cumsum()
ax = s.plot.line(color="red", figsize=(10,6))
ax.set_ylabel("Cases")
ax.set_xlabel("Days")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
plt.tight_layout()
plt.savefig("cases2.pdf")
```
