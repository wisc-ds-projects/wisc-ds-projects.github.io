```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.rcParams["font.size"] = 16
```

```python
s = pd.Series(np.random.randint(-3,10,100)).cumsum()
ax = s.plot.line(color="red", figsize=(5,3))
ax.set_ylabel("Cases")
ax.set_xlabel("Days")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
plt.tight_layout()
plt.savefig("cases.eps")
plt.savefig("cases.svg")
```
